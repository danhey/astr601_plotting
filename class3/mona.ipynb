```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Load image
img = mpimg.imread("data/mona.jpg")[:, :, 0]
```

```python
# Display image
plt.imshow(img, cmap="grey")
plt.axis("off")  # Hide axis
plt.show()
```

```python
plt.imshow(img, cmap="jet")
plt.axis("off")
plt.show()
```

```python
plt.imshow(img, cmap="viridis")
plt.axis("off")
plt.show()
```
