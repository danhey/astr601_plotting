```python
df["log_lum"] = np.log10(df["Luminosity [L_sun]"])
df["dist"] = 1 / df["Parallax [mas]"]
df["size"] = np.sqrt(df["Radius [R_sun]"]) * 10 + 50
```

```python
fig = px.scatter(
    df,
    x="Teff [K]",
    y="log_lum",
    size="size",
    color="dist",
    width=1000,
    height=1000,
    opacity=1,
    labels={"log_lum": r"log10 Luminosity"},
)

# Set x and y limits
fig.update_xaxes(range=[10000, 2000])
fig.update_yaxes(range=[-3, 6])
```
